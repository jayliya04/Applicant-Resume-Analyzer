```python
import pandas as pd
```

```python
df=pd.read_csv('resume_screening_job_dataset.csv')
```

```python
df.head()
```

```python
all_skills = (
    df['Skills']
    .str.split(';')
    .explode()
    .str.strip()
    .replace('', pd.NA)
    .dropna()
    .tolist()
)

print(all_skills)
```

```python
all_skills.extend(['agno','github','gitlab','langchain','langsmith','langgraph','vector databases','generative ai',"Machine Learning","Deep Learning","NLP","GenAI","Agentic AI","Vector Databases","NumPy","Pandas","Matplotlib","Seaborn","Scikit-learn","TensorFlow","Keras","Spacy","NLTK","Gensim","Fasttext","Flask","Streamlit","LLMs","RAG","Sentence Transformers","LangChain","Agno","ChromaDB","Fine Tuning","Prompt Engineering","Jupyter Notebook","Cursor","VS Code","Git"])
```

```python
normalized = [s.strip().lower() for s in all_skills]
```

```python
unique_skills = sorted(set(normalized))

print(unique_skills)
```
